```python
import pickle as pkl
import numpy as np

with open('../binary_opens.pkl','rb') as readfile:
  # ** x is a dictionary, with idx: [chems]
  x = pkl.load(readfile)
```

```python
with open('I_500.pkl', 'wb') as writefile:
  for i, id in enumerate(x.keys()):
    binary_odor = x[id]
    intensity_odor = [np.random.randint(1, 6) if x == 1 else x for x in binary_odor ]
    pkl.dump([intensity_odor, i], writefile)
```
