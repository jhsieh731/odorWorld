```python
import pickle as pkl
import numpy as np

with open('../binary_opens.pkl','rb') as readfile:
  # ** x is a dictionary, with idx: [chems]
  x = pkl.load(readfile)
```

```python
x.values()
```

```python
with open('data/I_500.pkl', 'wb') as writefile:
  for i, id in enumerate(x.keys()):
    binary_odor = x[id]
    intensity_odor = [np.random.randint(1, 6) if x == 1 else x for x in binary_odor ]
    pkl.dump([intensity_odor, i], writefile)
```

```python
import pickle as pkl
import csv

with open('data/I_500.pkl', 'rb') as readfile:
    with open('I_dict.csv', 'w') as writefile:
        writer = csv.writer(writefile)
        writer.writerow(['idx','chems'])
        offsets = []
        while True:
            try:
                offset = pkl.load(readfile)
                writer.writerow([str(offset[1]),str(offset[0])])
                offsets.append(offset)
            except EOFError:
                break
    print(offsets[0])
```
